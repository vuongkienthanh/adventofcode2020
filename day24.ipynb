```python
url = 'https://adventofcode.com/2020/day/24/input'
from helper import getdata
t = getdata(url)
from collections import defaultdict
from itertools import product
```

```python
class Map:
    def __init__(self):
        self.mem = defaultdict(lambda : 0)
    
    def run(self, line):
        e = lambda x,y,z: (x+1, y-1, z)
        w = lambda x,y,z: (x-1, y+1, z)
        ne = lambda x,y,z: (x+1, y, z-1)
        nw = lambda x,y,z: (x, y+1, z-1)
        se = lambda x,y,z: (x, y-1, z+1)
        sw = lambda x,y,z: (x-1, y, z+1)
        i=0
        x=y=z=0
        while i < len(line):
            if line[i] in 'ns':
                cmd = line[i:i+2]
                i += 1
            else:
                cmd = line[i]
            i += 1
            x,y,z = eval(f'{cmd}(x,y,z)')
        self.mem[(x,y,z)] = 1 - self.mem[(x,y,z)]
        
    def feed(self, data):
        for line in data:
            self.run(line)
        return sum(self.mem.values())
        
            
    def check(self, xi,yi,zi):
        res = sum([
            self.mem[(a,b,c)] for a,b,c in product(
                (xi-1,xi,xi+1),
                (yi-1,yi,yi+1),
                (zi-1,zi,zi+1)) if (sum([a,b,c])==0)&([a,b,c]!=[xi,yi,zi])
        ])
        if self.mem[(xi,yi,zi)] == 1:
            self.mem2[(xi,yi,zi)] = 1 - int((res==0)|(res>2))
        else:
            self.mem2[(xi,yi,zi)] = int(res==2)
    
    def loop_day(self, days):
        x,y,z = list(zip(*self.mem.keys()))
        r = max([
            abs(i) for i in 
                [min(x), max(x), min(y), max(y), min(z), max(z)]
        ]) + 1
        print('initial', r)
        for i in range(1, days+1):
            self.mem2 = self.mem.copy()
            for xi,yi,zi in product(range(-r, r+1),
                                    range(-r, r+1),
                                    range(-r, r+1)):
                if (xi+yi+zi)==0:
                    self.check(xi,yi,zi)
            
            print(f'day {i}', f'radius {r}', 'num blk',sum(self.mem2.values()))
            r += 1
            self.mem = self.mem2
```

## 1

```python
a = Map()
a.feed(t)
```

## 2

```python
a.loop_day(days=100)
```
